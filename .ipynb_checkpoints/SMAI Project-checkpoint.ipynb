```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# Load Images
pathp = 'C:/Users/Bhoomeendra/SMAI/faces/S'
imgPath = 'image_'
AllImages = list()
for fld in range(1,26):
    path=pathp+str(fld)
    singlePerson = list()
    for img in os.listdir(path):
        finalPath = os.path.join(path,img)
        cimg = cv2.imread(finalPath)
        cimg = cv2.cvtColor(cimg,cv2.COLOR_BGR2GRAY)
        cimg = cv2.resize(cimg,(448,296))
        singlePerson.append(cimg)
    AllImages.append(singlePerson)
print("Done")
```

```python
#Extrating Faces
AllFaces = list()
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
for i in range(len(AllImages)):
    for j in range(len(AllImages[i])):
        x,y,w,h = face_cascade.detectMultiScale(AllImages[i][j], 1.1, 5)
        face = AllImages[i][j][x:x+w,y:y+h]
        AllFaces.append(face/255)
plt.imshow(AllFaces[0],cmap='gray')
```

```python
#Calculating Matrix A
A = np.zeros((100,448*296))
for j in range(25):
    for i in range(4):# 4 image per Person
        flt = AllImages[j][i].flatten()
        m = np.mean(flt)
        for k in range(len(flt)):
            A[j*4+i][k]=flt[k]-m
```

```python
#Covariance Matrix
Cov = np.matmul(A,A.T)
```

```python
#Calculating Eigen Vector and Value
w , v = np.linalg.eig(Cov)
```

```python
#Sorting eVal_eganVec
eVal_eganVec = list()
for i in range(100):
    eVal_eganVec.append([w[i] , v[:,i]])
sortedeeV =sorted(eVal_eganVec, key=lambda x: x[0],reverse=True)
```

```python
#Converting eigenvector from m x1  to n^2 x 1
eigenVector = list()
for i in sortedeeV:
    eigenVector.append(np.matmul(A.T,i[1]))
```

```python
fig ,axs = plt.subplots(5,3,figsize=(15,18))
for i in range(5):
    for j in range(3):
        axs[i,j].set_title(str(3*i+j+1) + " EigenFaces ")
        axs[i,j].imshow(eigenVector[3*i+j].reshape(296,448),cmap = 'gray')
```

```python
def eigenRepresentation(img,eigenVector,dim = 100):
    imgReresentation = np.zeros((dim))
    for i in range(dim):
        imgReresentation[i] = np.dot(img,eigenVector[i])
    return imgReresentation/(130816)

allKnowFaces = list()
for j in range(100):
    flt = A[j,:]
    allKnowFaces.append(eigenRepresentation(flt,eigenVector))
```

```python
unKownFace = AllImages[5][5].flatten()
unKownFace = unKownFace - np.mean(unKownFace)
egnImg = eigenRepresentation(unKownFace.flatten(),eigenVector)
mindist = 100000000000
person= -1
for i in range(len(allKnowFaces)):
    if(mindist > np.sum((allKnowFaces[i]-egnImg)**2)):
        mindist=np.sum((allKnowFaces[i]-egnImg)**2)
        print(mindist)
        person= i       
print(person)  
```
